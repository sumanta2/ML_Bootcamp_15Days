```python
import matplotlib.pyplot as plt
import numpy as np
```

## Simple Plot

```python
x1 = np.linspace(0,5,10)
y1 = x1**2
plt.plot(x1,y1)

plt.title("Days Square Chart")

plt.xlabel('Days')
plt.ylabel("Days Squared")
# plt.show()
```

## Multiple Plots

```python
plt.subplot(1,2,1)

plt.plot(x1,y1,'r')

plt.subplot(1,2,2)
plt.plot(y1,x1,'g')
```

## Figure Objects

```python
fig1 = plt.figure(figsize=(5,4), dpi=100)

axes1 = fig1.add_axes([0.1,0.1,0.9,0.9])

axes1.set_xlabel("Days")
axes1.set_ylabel("Days Squared")
axes1.set_title("Figure Example")

axes1.plot(x1,y1, label='x/x^2')

axes1.plot(y1,x1, label='x^2/x')

axes1.legend(loc=0)

axes2  = fig1.add_axes([0.45,0.45,0.4,0.3])
axes2.set_xlabel("Days")
axes2.set_ylabel("Days Squared")
axes2.set_title("Small Figure Example")
axes2.plot(x1,y1,'r')

axes2.text(0,40, "Knowledge Shelf")
```

```python
fig1.savefig('plot.png')
```

## Working with dataframes

```python
import pandas as pd
```

```python
ice = pd.read_csv('icecreamsales.csv')
ice = ice.sort_values(by='Temperature')

np_arr = ice.values

x2 = np_arr[:,0]
y2 = np_arr[:,1]

fig2 = plt.figure(figsize=(6,4))
axes3 = fig2.add_axes([0,0,1,1])
axes3.set_xlabel("Temp")
axes3.set_ylabel("Ice Cream Sales")
axes3.set_title("Temp Vs. Sales")
axes3.plot(x2,y2)

axes3.annotate('Best for ice cream sales', xy=(83, 536), xytext=(60, 525),
            arrowprops=dict(facecolor='black', shrink=0.05),)

plt.bar(x2,y2)
```
