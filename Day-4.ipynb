## Pie Charts

```python
import random

fig1 = plt.figure(figsize=(8,5), dpi=100)
axes1 = fig1.add_axes([0.1,0.1,0.9,0.9])

types = ['Water', 'Normal', 'Flying', 'Grass', 'Psychic', 'Bug', 'Fire', 'Poison',
'Ground', 'Rock', 'Fighting', 'Dark', 'Steel', 'Electric', 'Dragon', 'Fairy',
'Ghost', 'Ice']

poke_num = [133, 109, 101, 98, 85, 77, 68, 66, 65, 60, 57, 54, 53, 51, 50, 50, 46, 40]

colors = []
for i in range(18):
    rgb = (random.uniform(0, .5), random.uniform(0, .5), random.uniform(0, .5))
    colors.append(rgb)
    

explode = [0] * 18
print(explode)
explode[0] = 0.4

x1, x2, x3= plt.pie(poke_num, explode=explode, labels = types, colors=colors, autopct='%1.0f%%', shadow=True,
       startangle=90, textprops=dict(color="w"))

plt.legend(x1, types, loc='right', bbox_to_anchor=(1,0,0.5,1))
```
